```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_palette('colorblind')
palette = sns.color_palette()

from cox.store import Store
```

```python
plt.rcParams['figure.facecolor'] = (1, 1, 1, 1)
```

```python
exps = ['fast_adv_l2_0p25', 'fast_adv_l2_0p5', 'fast_adv_l2_1p0']
exps_pgd = ['pgd_0p25', 'pgd_0p5', 'pgd_1p0']
es = [0.25, 0.5, 1.0]

for i, (e, e_pgd) in enumerate(zip(exps, exps_pgd)):
    s = Store('logs/', e)
    df = s['logs'].df
    df.head()
    ind = df.epoch <= 149
    plt.plot(df.epoch[ind], df.adv_prec1[ind], color=palette[i], label=f'$\epsilon$ = {es[i]}')
    
    s = Store('logs/', e_pgd)
    df = s['logs'].df
    ind = df.epoch <= 149
    plt.plot(df.epoch[ind], df.adv_prec1[ind], '--')
    
plt.legend()
plt.title('CIFAR-10, Dashed is PGD, Solid is Fast Adversarial')
plt.xlabel('Iterations')
plt.ylabel('Adversarial Accuracy')
```

```python
exps = ['fast_adv_l2_0p25', 'fast_adv_l2_0p5', 'fast_adv_l2_1p0']
exps_pgd = ['pgd_0p25', 'pgd_0p5', 'pgd_1p0']
es = [0.25, 0.5, 1.0]

for i, (e, e_pgd) in enumerate(zip(exps, exps_pgd)):
    s = Store('logs/', e)
    df = s['logs'].df
    df.head()
    ind = df.epoch <= 149    
    x1 = df.epoch[ind]
    y1 = df.adv_prec1[ind]
    
    s = Store('logs/', e_pgd)
    df = s['logs'].df
    ind = df.epoch <= 149
    x2 = df.epoch[ind]
    y2 = df.adv_prec1[ind]
    plt.plot(x1, -abs(y1-y2), label=f'$\epsilon$ = {es[i]}')
    
plt.legend()
plt.title('CIFAR-10')
plt.xlabel('Iterations')
plt.ylabel('Change in Adversarial Accuracy')
```

```python
exps = ['fast_adv_l2_0p25', 'fast_adv_l2_0p5', 'fast_adv_l2_1p0']
exps_pgd = ['pgd_0p25', 'pgd_0p5', 'pgd_1p0']
es = [0.25, 0.5, 1.0]

for i, (e, e_pgd) in enumerate(zip(exps, exps_pgd)):
    s = Store('logs/', e)
    df = s['logs'].df
    df.head()
    ind = df.epoch <= 149
    
    hours = (df.time - df.time[0])/60/60
    plt.plot(hours[ind], df.adv_prec1[ind], color=palette[i], label=f'$\epsilon$ = {es[i]}')
    
    s = Store('logs/', e_pgd)
    df = s['logs'].df
    ind = df.epoch <= 149
    hours = (df.time - df.time[0])/60/60
    plt.plot(hours[ind], df.adv_prec1[ind], '--')
    
plt.legend()
plt.title('CIFAR-10, Dashed is PGD, Solid is Fast Adversarial')
plt.xlabel('Training Hours')

plt.ylabel('Adversarial Accuracy')
```
